```python
AAPLOG.plot.line(y='Close',use_index=True)
plt.ylabel("PRICE")
plt.title('META PRICE OVER YEARS')
```

```python
AAPLOG['TOMORROW']=AAPLOG['Close'].shift(-1)
```

```python
AAPLOG.head()
```

```python
AAPLOG.tail()
```

```python

AAPLOG['Target'] = (AAPLOG['TOMORROW'] > AAPLOG['Close']).astype(int)
```

```python
AAPLOG
```

```python
AAPLOG.info()
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100,min_samples_split=100,random_state=42)
```

```python
x = AAPLOG.drop(columns=['Target'])
y=AAPLOG['Target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=43)
```

```python
RF.fit(x_train,y_train)
```

```python
y_pred = RF.predict(x_test)
```

```python

from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
score2 = accuracy_score(y_test,y_pred)
print("Accuracy:",score2)
cm2 = confusion_matrix(y_test,y_pred)
print(cm2)
cm3 = confusion_matrix(y_pred,y_test)
print(cm3)
print(classification_report(y_test,y_pred))
```
